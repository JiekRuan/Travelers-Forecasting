```python
import pandas
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn
import numpy

from sklearn.linear_model import LinearRegression
```

```python
path_csv = "D:/sklearn/linear_regression/number of travelers.csv"
dataset_df = pandas.read_csv(path_csv, sep=",")
```

# Data preparation
- Normalisation : standard
- train / test

```python
dataset_df.head()
```

```python
plt.figure(figsize=(15, 6))
plt.plot(dataset_df.index, dataset_df["passengers"])
plt.show()
```

```python
dataset_df["index_time"] = numpy.arange(1, len(dataset_df)+1, 1)
```

```python
dataset_mean, dataset_std = dataset_df["passengers"].mean(), dataset_df["passengers"].std()
dataset_df["passengers"] = (dataset_df["passengers"] - dataset_mean)/dataset_std
```

```python
dataset_df
```

```python
dataset_length = len(dataset_df)
index_split = int(dataset_length*0.75)
train_df = dataset_df[ : index_split]
test_df = dataset_df[index_split: ]
```

```python
x_train = train_df[["index_time"]].values
y_train = train_df[["passengers"]].values


x_test = test_df[["index_time"]].values
y_test = test_df[["passengers"]].values
```

# Entrainement
- Choix du modèle : regression linéaire
- instancier
- entraine à l'aide des données que l'on a préparées
- test

```python
model_passenger_prediction = LinearRegression()
```

```python
model_passenger_prediction.fit(x_train, y_train)
```

### indictions :
pour utiliser le modèle entrainé, regardez la document de predict

```python
dir(model_passenger_prediction)
```

```python
model_passenger_prediction.predict(y_train)
```

```python
y_pred = model_passenger_prediction.predict(x_test)
y_pred_train = model_passenger_prediction.predict(x_train)
```

```python
print(y_pred)
```

```python

y_train *= dataset_mean + dataset_std
y_test *= dataset_mean + dataset_std

y_pred *= dataset_mean + dataset_std
y_pred_train *= dataset_mean + dataset_std
```

```python
plt.figure(figsize=(15,6))
plt.scatter(x_test, y_test, label="verité terrain")
plt.plot(x_test, y_pred, label="forecast", color="r")

plt.legend()
plt.title('Comparaison du modèle et de la vérité terrain')
plt.xlabel('Mois')
plt.ylabel('Nombre de passagers', color="r")
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.scatter(x_train, y_train, label="verité terrain")
plt.plot(x_train, y_pred_train, label="forecast", color="r")

plt.legend()
plt.title('Comparaison du modèle et de la vérité terrain')
plt.xlabel('Mois')
plt.ylabel('Nombre de passagers', color="r")
plt.show()
```
